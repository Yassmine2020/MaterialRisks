### Elements between parentehsis

```python
import pdfplumber
import re

# Define the correct path to the PDF file
pdf_path = "mcs2024.pdf"

# Function to extract text between parentheses after the title and before the next heading
def extract_relevant_text(text):
    # Find the text between 'ABRASIVES (MANUFACTURED)' and 'Domestic Production and Use:'
    pattern = re.compile(r'ABRASIVES \(MANUFACTURED\)(.*)Domestic Production and Use:', re.DOTALL)
    match = pattern.search(text)
    if match:
        relevant_text = match.group(1)
        # Extract text within parentheses and remove parentheses
        parentheses_text = re.findall(r'\((.*?)\)', relevant_text)
        return parentheses_text
    return []

# Open the PDF file and extract text from page 33
with pdfplumber.open(pdf_path) as pdf:
    page = pdf.pages[33]  # Page 33 is index 32
    text = page.extract_text()

# Extract the relevant text
parentheses_text = extract_relevant_text(text)
print(parentheses_text)
```

### PDF table to lines (pdfplumber)

```python
import pdfplumber
import re

# Define the correct path to the PDF file
pdf_path = "mcs2024.pdf"

# Function to extract text between the specified headings
def extract_text_between_headings(text, start_heading, end_heading):
    # Find the text between start_heading and end_heading
    pattern = re.compile(re.escape(start_heading) + r'(.*?)' + re.escape(end_heading), re.DOTALL)
    match = pattern.search(text)
    if match:
        return match.group(1).strip()
    return ""

# Open the PDF file and extract text from page 33
with pdfplumber.open(pdf_path) as pdf:
    page = pdf.pages[46]  # Page 33 is index 32
    text = page.extract_text()

# Extract the text between 'ABRASIVES (MANUFACTURED)' and 'Domestic Production and Use:'
start_heading = 'Russia were revised based on company and Government reports.'
# start_heading = "World Smelter Production and Capacity"
end_heading = "World Resources"
relevant_text = extract_text_between_headings(text, start_heading, end_heading)

rt_lines = relevant_text.split('\n')
```

### PDF table to words (fitz)

```python
import fitz  # PyMuPDF

def read_pdf(file_path):
    # Open the PDF file
    document = fitz.open(file_path)
    
    # Create a list to hold the pages as strings
    pages = []
    
    # Iterate through each page
    for page_num in range(len(document)):
        # Get the page
        page = document.load_page(page_num)
        
        # Extract text from the page
        text = page.get_text("text")
        
        # Append the text to the pages list
        pages.append(text)
    
    return pages

# Example usage
file_path = 'mcs2024.pdf'
pdf_pages = read_pdf(file_path)

# Print the text of the first page
page = pdf_pages[46]
```

```python
rel_textpymu = extract_text_between_headings(page, start_heading, end_heading)

rt_words = rel_textpymu.split('\n')
```

### Text to csv

```python
res = []
for line in rt_lines:
    line_len = len(line)
    line_words = []
    words_lenght = 0
    for word in rt_words:
        # Strip the word to remove any leading/trailing whitespace and calculate its length
        word_len = len(word.strip())
        words_lenght += word_len
        if words_lenght > line_len:  # If the total length exceeds the line length, break
            break
        else:
            line_words.append(word.strip())  # Add the stripped word to the line_words
            if len(" ".join(line_words)) >= line_len:  # Check if the current words match the line length
                break
    res.append(line_words)
    # Remove the words that have been used in the current line
    rt_words = rt_words[len(line_words):]

res
```
